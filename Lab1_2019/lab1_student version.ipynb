### Α) Δημιουργία και οπτικοποίηση σύνθετων τύπων δικτύου

Κατασκευή και οπτικοποίηση της τοπολογίας REG για βαθμό πλέγματος $d = 4$

```python
d = 4

# για την κατασκευή του REG χρησιμοποιήστε την watts_strogatz_graph από το networkx. Επιλέξτε προσεκτικά τις παραμέτρους. 
REG = nx.watts_strogatz_graph(n, 4, 0)

# για την οπτικοποίηση, χρησιμοποιήστε την draw_circular από το networkx
nx.draw_circular(REG, node_size=100)
plt.show()
```

Κατασκευή και οπτικοποίηση της τοπολογίας RG(ER) για αριθμό συνδέσεων $M = 750$

```python
M = 750

# για την κατασκευή του RGER χρησιμοποιήστε την gnm_random_graph από το networkx
RGER =  nx.gnm_random_graph(n, M)

# για την οπτικοποίηση, χρησιμοποιήστε την draw_circular από το networkx
nx.draw_circular(RGER, node_size=100)
plt.show()
```

Κατασκευή και οπτικοποίηση της τοπολογίας RGG σε περιοχή $L\times L = 1$ (unit cube) και ακτίνα σύνδεσης $R=0.25$

```python
R = 0.25
L = 1
n = 100

# για την κατασκευή του RGG, χρησιμοποιήστε την random_geometric_graph από το networkx
RGG =  nx.random_geometric_graph(n, R, 2*L)

# για την οπτικοποίηση, χρησιμοποιήστε draw_network_edges, draw_network_nodes. 
# Για τη θέση των κόμβων στην τετράγωνη περιοχή, χρησιμοποιήστε την get_node_attributes από το networkx.
pos = nx.get_node_attributes(RGG, "pos")
# print(pos)
nx.draw_networkx_nodes(RGG, pos, node_size=100)
nx.draw_networkx_edges(RGG, pos)
plt.show()
```

Κατασκευή και οπτικοποίηση της τοπολογίας SF για βαθμό αρχικού πλέγματος $d=4$

```python
d = 4

# για την κατασκευή του SF, χρησιμοποιήστε την barabasi_albert_graph από το networkx.
SF = nx.barabasi_albert_graph(n, d)

# για την οπτικοποίηση, χρησιμοποιήστε την draw_circular από το networkx.
nx.draw_circular(SF, node_size=100)
```

Κατασκευή και οπτικοποίηση της τοπολογίας SW για βαθμό αρχικού πλέγματος $d=4$ και πιθανότητας ανασύνδεσης $p=0.3$

```python
d = 4
p = 0.3

# για την κατασκευή του SW, χρησιμοποιήστε την watts_strogatz_graph από το networkx.
SW = nx.watts_strogatz_graph(n, d, p)

# για την οπτικοποίηση, χρησιμοποιήστε την draw_circular από το networkx.
nx.draw_circular(SW, node_size=100)
```

##### i) Να υπολογιστεί η δύναμη του κάθε κόμβου

```python
#δίνεται η συνάρτηση strength που έχει ως όρισμα την τοπολογία που παράξατε στο ερώτημα Α:
def strength(topology):
    weights_list=[]
    for i in range(0,len(topology.adj)):
        a=[]
        for j in range (0,len(topology.adj[i].items())):
            d=list(topology.adj[i].values())[j]
            c=list(d.values())[0]
            a.append(c)
        weights_list.append(a)    
    node_weights = []
    for node in weights_list:
        node_weights.append(sum(node))
    return np.array(node_weights)

# np.array([v for (u, v) in nx.degree(G)])

# print(list(REG.adj[0].values())[0])
REG_strength = strength(REG)
print(REG_strength)
RGER_strength = strength(RGER)
RGG_strength=strength(RGG)
SF_strength=strength(SF)
SW_strength=strength(SW)
```

```python
## Προσοχή! Ο παρακάτω κώδικας παράγει τις τοπολογίες στις οποίες θα κάνετε αναλυτικούς υπολογισμούς.
## ΜΗΝ ΤΟΝ ΑΛΛΑΞΕΤΕ

plt.figure(1, figsize=(2,2))
plt.title('Topology 1')
G = nx.Graph()
pos = {0: (1, 1), 1: (0, 1), 2: (1, 0), 3: (2, 1), 4: (1, 2)}
G.add_nodes_from(pos.keys())
G.add_edges_from([(0,1),(0,2),(0,3),(0,4)])
labels = {}
labels[0] = r'$1$'
labels[1] = r'$2$'
labels[2] = r'$3$'
labels[3] = r'$4$'
labels[4] = r'$5$'
nx.draw_networkx_labels(G, pos, labels)
nx.draw(G,pos)

plt.figure(2,figsize=(2,2))
plt.title('Topology 2')
G = nx.Graph()
pos = {0: (1, 1), 1: (0, 0), 2: (2, 0), 3: (2, 2), 4: (0, 2)}
G.add_nodes_from(pos.keys())
G.add_edges_from([(0,1),(0,2),(0,3),(0,4),(1,2),(3,4)])
labels = {}
labels[0] = r'$1$'
labels[1] = r'$2$'
labels[2] = r'$3$'
labels[3] = r'$4$'
labels[4] = r'$5$'
nx.draw_networkx_labels(G, pos, labels)
nx.draw(G, pos)

plt.figure(3,figsize=(2,2))
plt.title('Topology 3')
G = nx.Graph()
pos = {0: (1, 1), 1: (0, 1), 2: (1, 0), 3: (2, 1), 4: (1, 2)}
G.add_nodes_from(pos.keys())
G.add_edges_from([(0,1),(0,2),(0,3),(0,4),(1,2),(2,3),(3,4),(4,1)])
labels = {}
labels[0] = r'$1$'
labels[1] = r'$2$'
labels[2] = r'$3$'
labels[3] = r'$4$'
labels[4] = r'$5$'
nx.draw_networkx_labels(G, pos, labels)
nx.draw(G,pos)
```

<h4>Αναλυτικός υπολογισμός συντελεστή ομαδοποίησης</h4>
<p>Για κάθε κόμβο χρησιμοποιούμε τον τύπο </p>

$$C_i = \frac{2L_i}{k_i ({k_i} - 1)}$$

<p>Όπου ${k_i}$ ο βαθμός του κόμβου $i$ και $L_i$ ο αριθμός των ακμών μεταξύ των $k_i$ γειτόνων του $i$ κόμβου</p>

<p><b>Topology 1</b><p>
    $$C_1 = \frac{2 * 0}{4 ({4} - 1)} = 0$$
    $$C_2 = \frac{2L_i}{1 ({k_i} - 1)}$$ 
    ο κόμβος 2 έχει μόνο 1 γείτονα (άρα και οι γείτονες του δεν έχουν μεταξύ τους ακμές), οπότε $C_2 = 0$
    Αντίστοιχα για τους κόμβους 3,4,5 της τοπολογίας 1
    
<p><b>Topology 2</b><p>
    $$C_1 = \frac{2 * 2}{4 ({4} - 1)} = \frac{1}{3}$$
    $$C_2 = \frac{2 * 1}{2 ({2} - 1)} = 1$$
    Αντίστοιχα με το ${C_2}$ και οι κόμβοι 3, 4, 5 έχουν clustering coefficient ίσο με 1

<h4>Αναλυτικός υπολογισμός κεντρικοτήτων</h4>
<p>Degree Centrality:
    $$C_{Di} = \frac{\sum_{j=1}^{n}a_{ij}}{n-1}$$
όπου $a_{ij}$ είναι 1 αν υπάρχει ακμή μεταξύ i και j, 0 αλλιώς
</p>

<p>Closeness Centrality:
    $$C(x) = \frac{n-1}{\sum_{y}d(y,x)}$$
όπου $d(y,x)$ είναι η απόσταση μεταξύ κόμβων y, x
</p>

<p>Betweenness Centrality:
    $$C_{B}(u) = \sum_{s,t\in{V}}\frac{\sigma(s,t|u)}{\sigma(s,t)}$$
όπου $\sigma(s,t)$ είναι το πλήθος των shortest paths μεταξύ s-t, $\sigma(s,t|u)$ είναι τα shortest paths εξ αυτών που περνούν από το $u$
<h5>Τοπολογία Α </h5>

**Degree Centrality:**
    $$C_{D1} = \frac{\sum_{j=1}^{n}a_{1j}}{n-1} = \frac{a_{12} + a_{13} + a_{14}}{3} = 1$$
    $$C_{D2} = \frac{\sum_{j=1}^{n}a_{ij}}{n-1} = \frac{a_{21}}{3} = \frac{1}{3}$$
Αντίστοιχα με τον 2 και οι κόμβοι 3,4

**Closeness Centrality:**
    $$C(1) = \frac{n-1}{\sum_{y}d(y,1)} = \frac{3}{1 + 1 + 1} = \frac{3}{3} = 1$$
    $$C(2) = \frac{n-1}{\sum_{y}d(y,2)} = \frac{3}{1 + 2 + 2} = \frac{3}{5}$$
Αντίστοιχα με κόμβο 2 οι κόμβοι 3 και 4

**Betweenness Centrality:**
    $$C_{B}(1) = \sum_{s,t\in{V}}\frac{\sigma(s,t|u)}{\sigma(s,t)}$$
όπου $\sigma(s,t)$ όλα τα μονοπάτια μεταξύ 2, 3 και 4, οπότε σύνολο 3. Άρα:
    $$C_{B}(1) = \sum_{s,t\in{V}}\frac{\sigma(s,t|u)}{3} = \frac{1}{3} + \frac{1}{3} + \frac{1}{3} = 1$$
    $$C_{B}(2) = 0$$
επειδή κανένα από τα shortest paths μεταξύ των υπολοίπων κόμβων δεν περνά από τον κόμβο 2!
Αντίστοιχα και για τους κόμβους 3,4
    

<h5>Τοπολογία B </h5>

**Degree Centrality:**
    $$C_{D1} = \frac{a_{12} + a_{13} + a_{14}}{3} = 1$$
Αντίστοιχα για όλους τους υπόλοιπους κόμβους

**Closeness Centrality:**
    $$C(1) = \frac{n-1}{\sum_{y}d(y,1)} = \frac{3}{1 + 1 + 1} = \frac{3}{3} = 1$$
Όλοι οι κόμβοι απέχουν 1 βήμα από κάθε άλλον στην τοπολογία, οπότε για κάθε κόμβο το closeness centrality είναι ${1}$

**Betweenness Centrality:**
    $$C_{B}(u) = \sum_{s,t\in{V}}\frac{\sigma(s,t|u)}{\sigma(s,t)}$$
Μπορούμε εύκολα να διακρίνουμε ότι για κάθε κόμβο $u$, τα shortest paths μεταξύ των υπολοίπων κόμβων $n - u$ δεν περνάνε από τον κόμβο $u$. Οπότε για όλους τους κόμβους της τοπολογίας το betweenness centrality είναι ίσο με $0$
